```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define hyperparameter grid
param_grid = {
    'penalty': ['l1', 'l2'],  # Regularization type
    'C': [0.1, 1.0, 10],  # Inverse regularization strength
    'solver': ['liblinear']  # Use liblinear for l1 or l2 penalties
}

# Set up Grid Search
grid_search = GridSearchCV(LogisticRegression(random_state=42), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X, y)

# Output the best parameters
print("Best Parameters for Logistic Regression:", grid_search.best_params_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Predictions with the best model
best_logistic_model = grid_search.best_estimator_
y_pred = best_logistic_model.predict(X)

# Print evaluation metrics
print("Logistic Regression Accuracy:", accuracy_score(y, y_pred))
print("Classification Report for Logistic Regression:\n", classification_report(y, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from imblearn.over_sampling import SMOTE
import pandas as pd

# Sample parameter grid with reduced options for faster search
param_grid = {
    'n_estimators': [100, 200],  # Fewer values for quicker tuning
    'max_depth': [10, 20],       # Reduced depth options
    'min_samples_split': [2, 5]  # Fewer split options
}

# Ensure all features are numeric and handle missing values in X and y
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)
y = y.fillna(0).astype(int)

# Step 1: Handle class imbalance using SMOTE (Oversampling the minority class)
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Step 2: Create a smaller sample for quick testing (optional)
X_sample = X_resampled.sample(frac=0.1, random_state=42)  # Take 10% of data
y_sample = y_resampled.loc[X_sample.index]

# Step 3: Set up RandomizedSearchCV for Random Forest
random_search_rf = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=5,  # Limits to 5 random combinations for speed
    cv=3,  # Reduce cross-validation folds
    scoring='accuracy',
    random_state=42,
    n_jobs=1  # Use single core for stability
)

# Step 4: Fit the RandomizedSearchCV on the sample data (for quick test) or full data
random_search_rf.fit(X_sample, y_sample)  # For quick testing
# random_search_rf.fit(X_resampled, y_resampled)  # Uncomment to run on full data

# Step 5: Output the best parameters and score
print("Best Parameters from Randomized Search:", random_search_rf.best_params_)
print("Best Score from Randomized Search:", random_search_rf.best_score_)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Get the best model from RandomizedSearchCV
best_rf_model = random_search_rf.best_estimator_

# Generate predictions on the resampled data
y_rf_pred = best_rf_model.predict(X_resampled)

# Print evaluation metrics
print("Random Forest Accuracy:", accuracy_score(y_resampled, y_rf_pred))
print("Random Forest Precision:", precision_score(y_resampled, y_rf_pred))
print("Random Forest Recall:", recall_score(y_resampled, y_rf_pred))
print("Random Forest F1 Score:", f1_score(y_resampled, y_rf_pred))
print("Classification Report for Random Forest:\n", classification_report(y_resampled, y_rf_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion matrix for the best Random Forest model
cm_rf = confusion_matrix(y_resampled, y_rf_pred)
ConfusionMatrixDisplay(cm_rf).plot()
plt.title("Confusion Matrix for Tuned Random Forest")
plt.show()
```
